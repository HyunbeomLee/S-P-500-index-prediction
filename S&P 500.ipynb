```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Desktop/sp500.csv")
df
```

```python
df['DateTime'] = pd.to_datetime(df.Date)
df_ordered = df.sort_values('DateTime', ascending=True)
df_ordered['index'] = range(0,df.shape[0],1)
df_ordered.set_index(['index'])
```

```python
df_ordered['date_after_april1_2018'] = df_ordered.DateTime > datetime(year=2018, month=4, day=1)

# Price(close) mean
ordered_close = df_ordered.Close
ordered_close_log = np.log(ordered_close)
price_mean_5day = ordered_close_log.rolling(5).mean().shift(1)
price_mean_365day = ordered_close_log.rolling(365).mean().shift(1)
price_mean_ratio = price_mean_5day/price_mean_365day

#Price(Close) Std
price_std_5day = ordered_close_log.rolling(5).std().shift(1)
price_std_365day = ordered_close_log.rolling(365).std().shift(1)
price_std_ratio = price_std_5day/price_std_365day
```

```python
#Volume mean
ordered_volume = df_ordered.Volume
ordered_volume_log = np.log(ordered_volume)
volume_mean_5day = ordered_volume_log.rolling(5).mean().shift(1)
volume_mean_365day = ordered_volume_log.rolling(365).mean().shift(1)
volume_mean_ratio = volume_mean_5day/volume_mean_365day

#Volume std
volume_std_5day = ordered_volume_log.rolling(5).std().shift(1)
volume_std_365day = ordered_volume_log.rolling(365).std().shift(1)
volume_std_ratio = volume_std_5day/volume_std_365day

```

```python
# Price(close) mean
df_ordered['price_mean_5day'] = price_mean_5day
df_ordered['price_mean_365day'] = price_mean_365day
df_ordered['price_mean_ratio'] = price_mean_ratio

#Price(Close) Std
df_ordered['price_std_5day'] = price_std_5day
df_ordered['price_std_365day'] = price_std_365day
df_ordered['price_std_ratio'] = price_std_ratio

#Volume(mean)
df_ordered['volume_mean_5day'] = volume_mean_5day
df_ordered['volume_mean_365day'] = volume_mean_365day
df_ordered['volume_mean_ratio'] = volume_mean_ratio

#Voume(std)
df_ordered['volume_std_5day'] = volume_std_5day
df_ordered['volume_std_365day'] = volume_std_365day
df_ordered['volume_std_ratio'] = volume_std_ratio
```

```python
df_new = df_ordered[df_ordered["DateTime"] > datetime(year=1951, month=1, day=2)]
df_no_NA = df_new.dropna(axis=0)

df_train = df_no_NA[df_no_NA['DateTime'] < datetime(year=2015, month=1, day=1)]
df_test = df_no_NA[df_no_NA['DateTime'] >= datetime(year=2015, month=1, day=1)]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
features = ['price_mean_5day', 'price_mean_365day', 'price_mean_ratio',
            'price_std_5day', 'price_std_365day', 'price_std_ratio',
            'volume_mean_5day', 'volume_mean_365day', 'volume_mean_ratio',
           'volume_std_5day','volume_std_365day', 'volume_std_ratio' ]
X = df_train[features]
X_test = df_test[features]
y = df_train.Close
y_test = df_test.Close
```

```python
model.fit(X, y)
pred = model.predict(X_test)
```

```python
MAE = sum(abs(pred - y_test))/len(pred)
print(MAE )
print(model.score(X, y))
```
